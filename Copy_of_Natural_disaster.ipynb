Data Cleaning

```python
#Only taking useful part of datetime
data1_df['declarationDate']=data1_df['declarationDate'].str.slice(0,10)
data1_df['incidentBeginDate']=data1_df['incidentBeginDate'].str.slice(0,10)
data1_df['incidentEndDate']=data1_df['incidentEndDate'].str.slice(0,10)
```

```python
#Replacing all "county" from the below column
data1_df['designatedAreaCounty']=data1_df['designatedArea'].str.replace('(County)', '')
```

```python
data1_df['designatedAreaCounty'] = data1_df['designatedAreaCounty'].str.split(' ').str[0]
```

```python
#Dropping columns not needed
data1_df=data1_df.drop(["femaDeclarationString","designatedArea","designatedArea","lastRefresh", "hash", "id", "disasterCloseoutDate", "lastIAFilingDate", "fipsStateCode", "fipsCountyCode", "placeCode"],axis=1)
```

```python
#Declaring schema
job_config = bigquery.LoadJobConfig(
    # Specify a (partial) schema. All columns are always written to the
    # table. The schema is used to assist in data type definitions.
    schema=[
        # Specify the type of columns whose type cannot be auto-detected. For
        # example the "title" column uses pandas dtype "object", so its
        # data type is ambiguous.
        bigquery.SchemaField("disasterNumber", bigquery.enums.SqlTypeNames.STRING),
        # Indexes are written if included in the schema by name.
        bigquery.SchemaField("state", bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField("declarationType", bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField("declarationDate", bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField("fyDeclared", bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField("incidentType", bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField("declarationTitle", bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField("ihProgramDeclared", bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField("iaProgramDeclared", bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField("paProgramDeclared", bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField("hmProgramDeclared", bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField("incidentBeginDate", bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField("incidentEndDate", bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField("declarationRequestNumber", bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField("designatedAreaCounty", bigquery.enums.SqlTypeNames.STRING),
       
    ], write_disposition="WRITE_APPEND")
```
